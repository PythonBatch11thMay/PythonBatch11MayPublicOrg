# **Classes and Object**

```python
#Creating a Class

class Dog:
    pass

dog1 = Dog()
dog2 = Dog()

#The Self Variable

class Dog:
    def __init__(self, name):
        self.name = name

dog1 = Dog("Buddy")
print(dog1.name)  # Output: Buddy

#Constructor
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

dog1 = Dog("Buddy", 3)
print(dog1.name, dog1.age)  # Output: Buddy 3

#Types of Variables
#Instance Variables: Variables that are unique to each instance.
#Class Variables: Variables that are shared among all instances of a class.

class Dog:
    species = "Canis familiaris"  # Class variable

    def __init__(self, name, age):
        self.name = name  # Instance variable
        self.age = age    # Instance variable

dog1 = Dog("Buddy", 3)
dog2 = Dog("Molly", 5)
print(dog1.species)  # Output: Canis familiaris
print(dog1.name)     # Output: Buddy


#Passing Members of One Class to Another Class

class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

class Car:
    def __init__(self, model, engine):
        self.model = model
        self.engine = engine

engine = Engine(200)
car = Car("Toyota", engine)
print(car.engine.horsepower)  # Output: 200


#Inner Classes

class Outer:
    class Inner:
        def __init__(self, value):
            self.value = value

outer = Outer()
inner = outer.Inner(10)
print(inner.value)  # Output: 10





class Dog:
    # Class attribute
    species = "Canis familiaris"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Instance method
    def speak(self):
        return f"{self.name} says Woof!"

    # Class method
    @classmethod
    def common_species(cls):
        return cls.species

# Using the class method
print(Dog.common_species())  # Output: Canis familiaris

# Creating an instance of the Dog class
obj = Dog("Mithun", 28)

# Using the instance method speak
print(obj.speak())  # Output: Mithun says Woof!


```

# **Inheritance**
Inheritance is a mechanism where a new class (subclass) is created based on an existing class (superclass), inheriting its attributes and methods. This allows the subclass to reuse code from the superclass and extend or modify its behavior.

```python
class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):  # Dog inherits from Animal
    def wag_tail(self):
        return "Dog wags tail"

# Creating instances
animal = Animal()
dog = Dog()

# Using inherited methods
print(animal.speak())  # Output: Animal speaks
print(dog.speak())     # Output: Animal speaks
print(dog.wag_tail())  # Output: Dog wags tail


# Animal is the superclass with a speak method.
# Dog is the subclass inheriting from Animal.
# Dog also has its own method wag_tail.
# Both Animal and Dog instances can call the speak method.


#1.Single Inheritance

class Animal:
    def move(self):
        return "Animal moves"

class Bird(Animal):
    def move(self):  # Method overridden in Bird
        return "Bird flies"

bird = Bird()
print(bird.move())  # Output: Bird flies

#2.Multiple Inheritance

class A:
    def method_a(self):
        return "Method A"

class B:
    def method_b(self):
        return "Method B"

class C(A, B):  # Inherits from both A and B
    def method_c(self):
        return "Method C"

c = C()
print(c.method_a())  # Output: Method A
print(c.method_b())  # Output: Method B
print(c.method_c())  # Output: Method C


#3.Multilevel Inheritance
class A:
    def method_a(self):
        return "Method A"

class B(A):  # Inherits from A
    def method_b(self):
        return "Method B"

class C(B):  # Inherits from B
    def method_c(self):
        return "Method C"

c = C()
print(c.method_a())  # Output: Method A
print(c.method_b())  # Output: Method B
print(c.method_c())  # Output: Method C



#4.Hierarchical Inheritance:
class Animal:
    def sound(self):
        return "Animal sound"

class Dog(Animal):
    def sound(self):  # Method overridden in Dog
        return "Dog barks"

class Cat(Animal):
    def sound(self):  # Method overridden in Cat
        return "Cat meows"

dog = Dog()
cat = Cat()

print(dog.sound())  # Output: Dog barks
print(cat.sound())  # Output: Cat meows
```

# **Polymorphism**
Polymorphism allows objects of different classes to be treated as objects of a common superclass. It allows methods to behave differently based on the object that calls them. This enables flexibility and extensibility in your code.

```python
class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):
    def speak(self):  # Method overridden in Dog
        return "Dog barks"

class Cat(Animal):
    def speak(self):  # Method overridden in Cat
        return "Cat meows"

# Creating instances
animal = Animal()
dog = Dog()
cat = Cat()

# Using overridden methods
print(animal.speak())  # Output: Animal speaks
print(dog.speak())     # Output: Dog barks
print(cat.speak())     # Output: Cat meows


#Both Dog and Cat classes have overridden the speak method from the superclass Animal.
#When speak is called on instances of Dog and Cat, it executes the method specific to each class.



# Superclass: Employee
# The Employee class serves as the superclass.It has common attributes like name and id, and
# a method calculate_salary() which is not implemented but serves as a placeholder for subclasses to override.

class Employee:
    def __init__(self, name, id):
        self.name = name
        self.id = id

    def calculate_salary(self):
        pass

# Subclass 1: HourlyEmployee
# The HourlyEmployee class is a subclass of Employee.It adds attributes hourly_rate and hours_worked specific to
# hourly employees. It overrides the calculate_salary() method to calculate the salary based on the hourly rate and hours worked.

class HourlyEmployee(Employee):
    def __init__(self, name, id, hourly_rate, hours_worked):
        super().__init__(name, id)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):  # Method overridden in HourlyEmployee
        return self.hourly_rate * self.hours_worked

#  Subclass 2: SalariedEmployee
# The SalariedEmployee class is another subclass of Employee. It adds an attribute salary specific to salaried employees.
# Similar to HourlyEmployee, it overrides the calculate_salary() method to return the fixed salary.

class SalariedEmployee(Employee):
    def __init__(self, name, id, salary):
        super().__init__(name, id)
        self.salary = salary

    def calculate_salary(self):  # Method overridden in SalariedEmployee
        return self.salary

hourly_emp = HourlyEmployee("John", 101, 15, 40)
salaried_emp = SalariedEmployee("Alice", 102, 50000)

print(hourly_emp.calculate_salary())  # Output: 600
print(salaried_emp.calculate_salary()) # Output: 50000

# Usage:
# We can create instances of HourlyEmployee and
# SalariedEmployee and call the calculate_salary() method on them.

# Inheritance allows HourlyEmployee and SalariedEmployee to reuse common attributes and methods from the Employee superclass.
# Polymorphism is demonstrated through the overridden calculate_salary() method in each subclass, where the same method behaves differently
# depending on the type of employee.
# The superclass provides a common interface (calculate_salary()) that allows subclasses to provide their own implementation suited to their
# specific type of employee.






#Code Reusability: Inheritance allows you to reuse code from existing classes in new ones.
#Flexibility: Polymorphism allows objects to be treated as instances of a superclass, promoting code flexibility and extensibility.



class Bird:
    def fly(self):
        print("Flying")

class Airplane:
    def fly(self):
        print("Flying in the sky")

# Example usage:
for thing in [Bird(), Airplane()]:
    thing.fly()
# Output:
# Flying
# Flying in the sky
```
